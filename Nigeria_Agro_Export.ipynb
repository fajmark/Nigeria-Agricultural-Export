```python
# Importing Libraries
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
# Define products and European countries
products = ['Cocoa', 'Palm Oil', 'Rubber', 'Cashew', 'Sesame', 'Ginger', 'Cassava', 'Plantain']
european_countries = ['Germany', 'France', 'Italy', 'Spain', 'Netherlands', 'Belgium', 'Switzerland', 'Sweden', 'Austria', 'Denmark']
```

```python
# Sample real company names (replace with actual company names)
company_names = ['Agro Export Nigeria Ltd', 'Farmgate Nigeria Limited', 'Greenfield Agro Exporters Nigeria Limited', 
                'Nigeria Agro Export Company', 'Golden Farms Nigeria Limited', 'Prime Agro Exports Nigeria Limited', 
                'Agriplus Nigeria Limited', 'Nigerian Export Promotion Council (NEPC)', 'Solid Agro Nigeria Limited', 
                'Agro Allied Exporters Nigeria Limited']
```

```python
# Define date range
start_date = datetime(2020, 1, 1)
end_date = datetime(2023, 12, 31)
```

```python
# Generate random data
data = []
for _ in range(1000): # Assuming 1000 exports
    product = random.choice(products)
    company = random.choice(company_names)
    export_country = random.choice(european_countries)
    date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    units_sold = random.randint(100, 1000)
    cogs = round(random.uniform(10000, 50000), 2)
    profit = round(random.uniform(2000, 10000), 2)
    export_value = round(units_sold * cogs, 2)
    destination_port = random.choice(['Lagos','Lagos', 'Lagos','Lagos','Lagos', 'Port Harcourt', 'Calabar', 'Warri'])
    transportation_mode = 'Sea'
    data.append([product, company, export_country, date, units_sold, cogs, profit, export_value, destination_port, transportation_mode])
```

```python
# Create DataFrame
df = pd.DataFrame(data, columns=['Product Name', 'Company', 'Export Country', 'Date', 'Units Sold', 'COGS', 'Profit', 'Export Value', 'Destination Port', 'Transportation Mode'])
```

```python
# Save DataFrame to CSV file
df.to_csv('nigeria_agricultural_exports.csv', index=False)
```
